Defining Parameters

```python
NUM_CUSTOMERS = 50

INDUSTRIES = [
    'Electric Utility',
    'Gas Utility', 
    'Water Utility',
    'Telecom',
    'Railways'
]

print(f"Generating data for {NUM_CUSTOMERS} customers")
print(f"Industries: {INDUSTRIES}")
```

Generating Each column

```python
customer_ids = [f'C{str(i).zfill(3)}' for i in range(1, NUM_CUSTOMERS + 1)]

customer_names = [fake.company() for _ in range(NUM_CUSTOMERS)]

industries = np.random.choice(INDUSTRIES, size=NUM_CUSTOMERS)

contract_values = np.random.choice(
    [50000, 75000, 100000, 150000, 200000],
    size=NUM_CUSTOMERS
)

logins_last_30_days = np.random.randint(0, 30, size=NUM_CUSTOMERS)

feature_adoption_rate = np.round(
    np.random.uniform(0.1, 1.0, size=NUM_CUSTOMERS), 2
)

open_tickets = np.random.randint(0, 10, size=NUM_CUSTOMERS)

days_since_last_login = np.random.randint(1, 60, size=NUM_CUSTOMERS)

contract_renewal_days = np.random.randint(10, 365, size=NUM_CUSTOMERS)

ai_acceptance_rate = np.round(
    np.random.uniform(0.2, 1.0, size=NUM_CUSTOMERS), 2
)

print("All columns generated successfully")
```

Assemble Dataframe

```python
df = pd.DataFrame({
    'customer_id'           : customer_ids,
    'customer_name'         : customer_names,
    'industry'              : industries,
    'contract_value'        : contract_values,
    'logins_last_30_days'   : logins_last_30_days,
    'feature_adoption_rate' : feature_adoption_rate,
    'open_tickets'          : open_tickets,
    'days_since_last_login' : days_since_last_login,
    'contract_renewal_days' : contract_renewal_days,
    'ai_acceptance_rate'    : ai_acceptance_rate
})

print(f"Dataframe shape: {df.shape}")
df.head(10)
```
